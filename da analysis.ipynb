```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取数据
df = pd.read_csv('user_personalized_features.csv', index_col=0)
print("数据形状:", df.shape)
print("\n数据前5行:")
print(df.head())

print("\n" + "任务1: 用户画像分析")

# 1.1 基础特征分布
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# 年龄分布
axes[0, 0].hist(df['Age'], bins=20, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('年龄分布')
axes[0, 0].set_xlabel('年龄')
axes[0, 0].set_ylabel('人数')

# 性别分布
gender_counts = df['Gender'].value_counts()
axes[0, 1].bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
axes[0, 1].set_title('性别分布')
for i, v in enumerate(gender_counts.values):
    axes[0, 1].text(i, v, str(v), ha='center', va='bottom')

# 地理位置分布
location_counts = df['Location'].value_counts()
axes[0, 2].pie(location_counts.values, labels=location_counts.index, autopct='%1.1f%%')
axes[0, 2].set_title('地理位置分布')

# 收入分布
axes[1, 0].hist(df['Income'], bins=20, edgecolor='black', alpha=0.7, color='green')
axes[1, 0].set_title('收入分布')
axes[1, 0].set_xlabel('收入')
axes[1, 0].set_ylabel('人数')

# 兴趣分布
interest_counts = df['Interests'].value_counts().head(10)
axes[1, 1].barh(interest_counts.index, interest_counts.values)
axes[1, 1].set_title('Top 10 兴趣分布')

# 订阅情况
sub_counts = df['Newsletter_Subscription'].value_counts()
axes[1, 2].pie(sub_counts.values, labels=['订阅' if x else '未订阅' for x in sub_counts.index], 
               autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
axes[1, 2].set_title('邮件订阅情况')

plt.tight_layout()
plt.savefig('user_profile_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# 1.2 订阅与非订阅用户消费对比
subscription_stats = df.groupby('Newsletter_Subscription').agg({
    'Total_Spending': 'mean',
    'Purchase_Frequency': 'mean',
    'Average_Order_Value': 'mean'
}).round(2)
print("\n订阅vs未订阅用户消费对比:")
print(subscription_stats)

print("\n" + "任务2: 消费行为分析")

# 2.1 用户价值分层
# 根据Total_Spending分位数划分
spending_quantiles = df['Total_Spending'].quantile([0.33, 0.67])
df['Spending_Level'] = pd.cut(df['Total_Spending'], 
                               bins=[-np.inf, spending_quantiles.iloc[0], 
                                     spending_quantiles.iloc[1], np.inf],
                               labels=['低价值', '中价值', '高价值'])

print("\n各价值层级用户数量:")
print(df['Spending_Level'].value_counts())

# 2.2 购买频率分析
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 购买频率分布
axes[0].hist(df['Purchase_Frequency'], bins=10, edgecolor='black', alpha=0.7)
axes[0].set_title('购买频率分布')
axes[0].set_xlabel('购买频率')
axes[0].set_ylabel('人数')

# 不同价值层级的购买频率
sns.boxplot(data=df, x='Spending_Level', y='Purchase_Frequency', ax=axes[1])
axes[1].set_title('不同价值层级用户的购买频率')

# 不同品类的平均消费
category_spending = df.groupby('Product_Category_Preference')['Total_Spending'].mean().sort_values()
axes[2].barh(category_spending.index, category_spending.values)
axes[2].set_title('各品类平均消费')

plt.tight_layout()
plt.savefig('consumption_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# 2.3 各品类的详细消费分析
category_analysis = df.groupby('Product_Category_Preference').agg({
    'Total_Spending': ['mean', 'median', 'std'],
    'Purchase_Frequency': 'mean',
    'Average_Order_Value': 'mean',
    'User_ID': 'count'
}).round(2)
print("\n各品类详细消费分析:")
print(category_analysis)

print("\n" + "任务3: 用户活跃度分析")

# 3.1 定义活跃度标签
df['Login_Status'] = pd.cut(df['Last_Login_Days_Ago'], 
                             bins=[-np.inf, 7, 30, np.inf],
                             labels=['活跃用户', '沉默用户', '流失风险'])

print("\n用户活跃状态分布:")
print(df['Login_Status'].value_counts())
print("\n百分比:")
print(df['Login_Status'].value_counts(normalize=True).round(3) * 100)

# 3.2 活跃度与消费行为的关系
login_status_analysis = df.groupby('Login_Status').agg({
    'Total_Spending': 'mean',
    'Purchase_Frequency': 'mean',
    'Time_Spent_on_Site_Minutes': 'mean',
    'Pages_Viewed': 'mean'
}).round(2)
print("\n不同活跃状态用户的消费行为:")
print(login_status_analysis)

# 3.3 可视化活跃度分析
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 最近登录时间分布
axes[0, 0].hist(df['Last_Login_Days_Ago'], bins=30, edgecolor='black', alpha=0.7)
axes[0, 0].axvline(x=7, color='green', linestyle='--', label='7天')
axes[0, 0].axvline(x=30, color='red', linestyle='--', label='30天')
axes[0, 0].set_title('最近登录天数分布')
axes[0, 0].set_xlabel('天数')
axes[0, 0].set_ylabel('人数')
axes[0, 0].legend()

# 站内时间 vs 浏览页数
axes[0, 1].scatter(df['Time_Spent_on_Site_Minutes'], df['Pages_Viewed'], 
                   alpha=0.5, c=df['Total_Spending'], cmap='viridis')
axes[0, 1].set_xlabel('站内时间(分钟)')
axes[0, 1].set_ylabel('浏览页数')
axes[0, 1].set_title('站内行为散点图(颜色=总消费)')
plt.colorbar(axes[0, 1].collections[0], ax=axes[0, 1])

# 不同活跃状态的消费对比
sns.boxplot(data=df, x='Login_Status', y='Total_Spending', ax=axes[1, 0])
axes[1, 0].set_title('不同活跃状态用户消费对比')

# 不同活跃状态的站内时间
sns.boxplot(data=df, x='Login_Status', y='Time_Spent_on_Site_Minutes', ax=axes[1, 1])
axes[1, 1].set_title('不同活跃状态用户站内时间对比')

plt.tight_layout()
plt.savefig('activity_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "任务4: RFM用户分层")

# 4.1 RFM评分计算
# R: 最近登录时间 (越小越好)
df['R_Score'] = pd.qcut(df['Last_Login_Days_Ago'], q=4, labels=[4, 3, 2, 1])

# F: 购买频率 (越大越好)
df['F_Score'] = pd.qcut(df['Purchase_Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])

# M: 总消费 (越大越好)
df['M_Score'] = pd.qcut(df['Total_Spending'].rank(method='first'), q=4, labels=[1, 2, 3, 4])

# 转换为数值型
df['R_Score'] = df['R_Score'].astype(int)
df['F_Score'] = df['F_Score'].astype(int)
df['M_Score'] = df['M_Score'].astype(int)

# 4.2 RFM总分
df['RFM_Score'] = df['R_Score'] + df['F_Score'] + df['M_Score']

# 4.3 定义用户层级
def define_rfm_segment(row):
    if row['R_Score'] >= 3 and row['F_Score'] >= 3 and row['M_Score'] >= 3:
        return '高价值用户'
    elif row['R_Score'] >= 3 and row['F_Score'] >= 2 and row['M_Score'] >= 2:
        return '潜力用户'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3 and row['M_Score'] >= 3:
        return '重要挽留用户'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2 and row['M_Score'] <= 2:
        return '低价值用户'
    elif row['R_Score'] >= 3 and row['F_Score'] <= 2 and row['M_Score'] <= 2:
        return '新用户'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3 and row['M_Score'] <= 2:
        return '高频低消用户'
    else:
        return '一般用户'

df['RFM_Segment'] = df.apply(define_rfm_segment, axis=1)

# 4.4 RFM分层结果
rfm_summary = df['RFM_Segment'].value_counts()
print("\nRFM用户分层结果:")
print(rfm_summary)
print("\n各层级占比:")
print((rfm_summary / len(df) * 100).round(2))

# 4.5 RFM各层级的消费特征
rfm_segment_analysis = df.groupby('RFM_Segment').agg({
    'Total_Spending': ['mean', 'median', 'sum'],
    'Purchase_Frequency': 'mean',
    'Average_Order_Value': 'mean',
    'Last_Login_Days_Ago': 'mean',
    'User_ID': 'count'
}).round(2)
print("\nRFM各层级详细分析:")
print(rfm_segment_analysis)

# 4.6 RFM可视化
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# RFM层级分布
rfm_counts = df['RFM_Segment'].value_counts()
axes[0, 0].bar(rfm_counts.index, rfm_counts.values)
axes[0, 0].set_title('RFM用户层级分布')
axes[0, 0].tick_params(axis='x', rotation=45)

# 各层级平均消费
segment_spending = df.groupby('RFM_Segment')['Total_Spending'].mean().sort_values()
axes[0, 1].barh(segment_spending.index, segment_spending.values)
axes[0, 1].set_title('各层级平均消费')

# RFM评分分布热力图
rfm_pivot = df.pivot_table(index='R_Score', columns='F_Score', values='M_Score', aggfunc='mean')
sns.heatmap(rfm_pivot, annot=True, cmap='YlOrRd', ax=axes[1, 0])
axes[1, 0].set_title('RFM评分矩阵')

# 各层级的最近登录时间
sns.boxplot(data=df, x='RFM_Segment', y='Last_Login_Days_Ago', ax=axes[1, 1])
axes[1, 1].set_title('各层级最近登录时间对比')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('rfm_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "任务5: 个性化推荐优化建议")

# 5.1 兴趣与购买品类的一致性分析
df['Interest_Category_Match'] = df['Interests'] == df['Product_Category_Preference']
match_rate = df['Interest_Category_Match'].mean() * 100
print(f"\n兴趣与购买品类匹配率: {match_rate:.2f}%")

# 查看不匹配的用户特征
mismatch_users = df[~df['Interest_Category_Match']]
print(f"\n不匹配用户数量: {len(mismatch_users)}")
print("\n不匹配用户的兴趣分布:")
print(mismatch_users['Interests'].value_counts(normalize=True).head())

# 5.2 兴趣vs实际购买交叉分析
interest_category_cross = pd.crosstab(df['Interests'], df['Product_Category_Preference'], 
                                       normalize='index') * 100
print("\n兴趣vs实际购买品类转化率(%):")
print(interest_category_cross.round(2))

# 5.3 可视化推荐机会
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 匹配vs不匹配用户的消费对比
match_comparison = df.groupby('Interest_Category_Match').agg({
    'Total_Spending': 'mean',
    'Purchase_Frequency': 'mean'
}).round(2)
match_comparison.plot(kind='bar', ax=axes[0])
axes[0].set_title('匹配vs不匹配用户的消费行为对比')
axes[0].set_xlabel('兴趣与购买是否匹配')
axes[0].legend(['平均消费', '购买频率'])

# 兴趣到品类的转化热力图
sns.heatmap(interest_category_cross, annot=True, fmt='.1f', cmap='YlOrRd', ax=axes[1])
axes[1].set_title('兴趣到购买品类的转化率(%)')

plt.tight_layout()
plt.savefig('recommendation_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# 5.4 推荐策略建议
print("\n推荐策略建议:")
print("""
1. 基于兴趣的推荐:
   - 对兴趣为Sports但购买Books的用户，推荐体育类书籍
   - 对兴趣为Technology但购买Apparel的用户，推荐科技主题服装
   
2. 交叉销售机会:
   - 购买Electronics的用户可推荐Technology相关商品
   - 购买Fashion的用户可推荐Apparel品类

3. 个性化邮件营销:
   - 根据用户兴趣推送相关品类的新品和促销
   - 对匹配度高的用户加强同类商品推荐
""")

print("\n" + "任务6: A/B测试建议")

# 6.1 识别适合A/B测试的用户群
# 为A/B测试准备用户分组
df['Test_Group'] = np.random.choice(['A', 'B'], size=len(df), p=[0.5, 0.5])

# 6.2 定义不同的A/B测试方案
print("\n建议的A/B测试方案:")

# 方案1: 邮件频率测试
print("\n【方案1】邮件推送频率测试")
print("目标用户: 订阅邮件的用户")
print("A组: 每周推送1次邮件")
print("B组: 每周推送3次邮件")
print("衡量指标: 打开率、点击率、转化率、7日内购买金额")

# 方案2: 个性化推荐测试
print("\n【方案2】推荐算法效果测试")
print("目标用户: 兴趣与购买不匹配的用户")
print("A组: 基于兴趣推荐")
print("B组: 基于历史购买推荐")
print("衡量指标: 推荐点击率、加购率、购买转化率")

# 方案3: 高价值用户运营测试
high_value_users = df[df['RFM_Segment'] == '高价值用户']
print(f"\n【方案3】高价值用户专属优惠测试")
print(f"目标用户: {len(high_value_users)}名高价值用户")
print("A组: 满减优惠券")
print("B组: 折扣优惠券") 
print("衡量指标: 优惠券使用率、客单价提升、复购率")

# 6.3 A/B测试用户分组导出
test_groups = df[['User_ID', 'RFM_Segment', 'Interests', 'Test_Group']].copy()
test_groups.to_csv('ab_test_groups.csv', index=False)
print("\nA/B测试分组已导出至 'ab_test_groups.csv'")

print("\n分析报告摘要")

report_summary = f"""
【用户画像摘要】
- 总用户数: {len(df)} 人
- 平均年龄: {df['Age'].mean():.1f} 岁
- 性别比例: {df['Gender'].value_counts(normalize=True).to_dict()}
- 邮件订阅率: {df['Newsletter_Subscription'].mean()*100:.1f}%

【消费行为摘要】
- 平均总消费: {df['Total_Spending'].mean():.2f}
- 平均购买频率: {df['Purchase_Frequency'].mean():.2f}
- 平均客单价: {df['Average_Order_Value'].mean():.2f}
- 高价值用户占比: {(df['Spending_Level']=='高价值').mean()*100:.1f}%

【活跃度摘要】
- 活跃用户占比: {(df['Login_Status']=='活跃用户').mean()*100:.1f}%
- 流失风险用户占比: {(df['Login_Status']=='流失风险').mean()*100:.1f}%
- 平均站内时间: {df['Time_Spent_on_Site_Minutes'].mean():.1f} 分钟
- 平均浏览页数: {df['Pages_Viewed'].mean():.1f} 页

【RFM分层摘要】
- 高价值用户: {(df['RFM_Segment']=='高价值用户').mean()*100:.1f}%
- 潜力用户: {(df['RFM_Segment']=='潜力用户').mean()*100:.1f}%
- 重要挽留用户: {(df['RFM_Segment']=='重要挽留用户').mean()*100:.1f}%
- 低价值用户: {(df['RFM_Segment']=='低价值用户').mean()*100:.1f}%

【推荐优化机会】
- 兴趣-购买匹配率: {match_rate:.1f}%
- 待优化用户数: {len(mismatch_users)} 人
"""

print(report_summary)

# 保存分析结果到Excel
with pd.ExcelWriter('user_analysis_report.xlsx') as writer:
    df.to_excel(writer, sheet_name='原始数据_带标签', index=False)
    rfm_segment_analysis.to_excel(writer, sheet_name='RFM分层分析')
    category_analysis.to_excel(writer, sheet_name='品类分析')
    subscription_stats.to_excel(writer, sheet_name='订阅对比')
    login_status_analysis.to_excel(writer, sheet_name='活跃度分析')
    interest_category_cross.to_excel(writer, sheet_name='兴趣品类交叉')

print("\n分析报告已保存至 'user_analysis_report.xlsx'")
print("可视化图片已保存: user_profile_analysis.png, consumption_analysis.png, activity_analysis.png, rfm_analysis.png, recommendation_analysis.png")
```
